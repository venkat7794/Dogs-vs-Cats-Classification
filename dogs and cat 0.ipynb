# Import necessary libraries

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
import matplotlib.pyplot as plt
import numpy as np
```

# Set the dimensions of the input images

```python
img_height, img_width = 200, 200
```

# Define the directories for the training and validation sets

```python
train_dir = '/content/images/train'
validation_dir = '/content/images/validation'
```


# Define the data generators for the training and validation sets

```python
train_datagen = ImageDataGenerator(
    rescale=1./255,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True
)

validation_datagen = ImageDataGenerator(rescale=1./255)

train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=(img_height, img_width),
    batch_size=32,
    class_mode='binary'
)

validation_generator = validation_datagen.flow_from_directory(
    validation_dir,
    target_size=(img_height, img_width),
    batch_size=32,
    class_mode='binary'
)
```

# Define the CNN model

```python
model = Sequential()
model.add(Conv2D(32, (3, 3), activation='relu', kernel_initializer='he_uniform', padding='same', input_shape=(img_height, img_width, 3)))
model.add(MaxPooling2D((2, 2)))
model.add(Conv2D(64, (3, 3), activation='relu', kernel_initializer='he_uniform', padding='same'))
model.add(MaxPooling2D((2, 2)))
model.add(Conv2D(128, (3, 3), activation='relu', kernel_initializer='he_uniform', padding='same'))
model.add(MaxPooling2D((2, 2)))
model.add(Flatten())
model.add(Dense(128, activation='relu', kernel_initializer='he_uniform'))
model.add(Dense(1, activation='sigmoid'))
```

# Compile the model

```python
opt = tf.keras.optimizers.SGD(lr=0.001, momentum=0.9)
model.compile(optimizer=opt, loss='binary_crossentropy', metrics=['accuracy'])
```


# Train the model

```python
history = model.fit(
    train_generator,
    epochs=10,
    validation_data=validation_generator
)
```


# Make a prediction on a single image

```python
import tensorflow as tf

img_path = '/content/images/train/cats/Image_1.jpg'
img = tf.io.read_file(img_path)
# Decode the image file into a tensor
img = tf.image.decode_jpeg(img, channels=3)  # Assuming JPEG format, adjust channels if needed
img = tf.image.resize(img, (img_height, img_width))
img = tf.image.convert_image_dtype(img, tf.float32)
img = tf.expand_dims(img, 0)
prediction = model.predict(img)
```

# Visualize layer outputs

```python
layer_outputs = model.layers[0].output
model_with_layer_outputs = tf.keras.models.Model(inputs=model.inputs, outputs=layer_outputs)
output = model_with_layer_outputs.predict(img)

# Iterate through feature maps and display them individually
for i in range(output.shape[-1]):
    plt.imshow(output[0, :, :, i], cmap='gray')  # Use 'gray' colormap for individual feature maps
    plt.title(f"Feature Map {i+1}")
    plt.show()
```

# Visualize the output of a single convolution filter

```python
filter_index = 0  # Index of the filter you want to visualize
conv_layer = model.layers[0]
filter = conv_layer.weights[0].numpy()  # Get the weights of the convolutional layer

# Iterate through and display each channel of the filter
for i in range(filter.shape[-1]):
    plt.imshow(filter[:, :, filter_index, i], cmap='gray')  # Select the filter and channel
    plt.title(f"Filter Channel {i+1}")
    plt.show()
```

```python
first_conv_layer = model.layers[1].output
model_with_first_conv_layer = tf.keras.models.Model(inputs=model.inputs, outputs=first_conv_layer)
output = model_with_first_conv_layer.predict(img)
for i in range(output.shape[-1]):
    plt.imshow(output[0, :, :, i], cmap='gray')  # Display each feature map
    plt.title(f"Feature Map {i+1}")
    plt.show()
```

# Visualize the second convolution layer output

```python
second_conv_layer = model.layers[3].output
model_with_second_conv_layer = tf.keras.models.Model(inputs=model.inputs, outputs=second_conv_layer)
output = model_with_second_conv_layer.predict(img)

# Iterate through and display each channel of the output
for i in range(output.shape[-1]):
    plt.imshow(output[0, :, :, i], cmap='gray')  # Select the channel
    plt.title(f"Feature Map {i+1}")
    plt.show()
```

# Model performance on unseen data

```python
test_dir = '/content/images'
test_datagen = ImageDataGenerator(rescale=1./255)
test_generator = test_datagen.flow_from_directory(
    test_dir,
    target_size=(img_height, img_width),
    batch_size=32,
    class_mode='binary'
)
```

```python
test_loss, test_acc = model.evaluate(test_generator)
print("Test Accuracy: ", test_acc)
print("Test Loss: ", test_loss)
```
